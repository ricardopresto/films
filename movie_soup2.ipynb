```python


def format_number(s):
    if s[0] != '$': return []
    n = re.search('[0-9]+([,.][0-9]+)*', s).group()
    if 'million' in s:
        if '.' in n:
            dec = len(n) - n.index('.') - 1
            n = n.replace('.', '') + '0' * (6 - dec)
        else: n = n + '000000'
    if 'billion' in s:
        if '.' in n:
            dec = len(n) - n.index('.') - 1
            n = n.replace('.', '') + '0' * (9 - dec)
        else: n = n + '000000000'
    if ',' not in n:
        n = '{0:,.0f}'.format(int(n))
    return n


def get_budget(number):
    res = []
    if info.find('th', string=number) is not None:
        res = info.find('th', string=number).parent.stripped_strings
    if res == []: return res
    try:
        return format_number(list(res)[1:][0])
    except:
        return []

```
